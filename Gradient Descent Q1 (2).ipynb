```python
import numpy as np
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
```

```python
def step_grad(X,Y,alpha,theta):
    theta_slope = np.zeros(X.shape[1])
    M= X.shape[0]
    for i in range(M):
        theta_slope +=(-2/M)*(Y[i] - ( X[i]*theta).sum() )*X[i] 
    new_theta = theta - (theta_slope*alpha)
    return new_theta
```

```python
def cost(X,Y,theta):
    totalcost = 0
    M,N = X.shape
    for i in range(M):
        theta_val = 0 
        for j in range(N):
            theta_val += (theta[j]*X[i][j])
        totalcost += (1/M)*((Y - theta_val)**2)
    return totalcost
```

```python
def gd(X,Y,alpha, num):
    theta = np.zeros(X.shape[1])
    for i in range(num):
        theta = step_grad(X,Y,alpha,theta)
    return theta
```
